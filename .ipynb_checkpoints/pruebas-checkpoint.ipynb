```python
#algo piola para utilizar para graficar es realizar una función como la que sigue:
def graficar(ax, datox, datoy, param_dicc):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    datox : array
       The x data

    datoy : array
       The y data

    param_dicc : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(datox, datoy, **param_dicc)
    return out

#np.random.normal devuelve valores aleatorios en base a una distribución normal cuya mediana es 0 (loc)
#se le puede fijar los numeros random que va tomando en ejecuciones sucesivas
#np.random.seed(20)
print("random normal desviacion estandar 5 (scale) -> {}".format(np.random.normal(5)))
print("random normal size=5 -> {}".format(np.random.normal(size=5)))
print("random normal size=(2,3) crea una matriz 2x3-> {}".format(np.random.normal(size=(2,3))))

#se le puede asignar un valor de mediana mediante loc:
print("random normal mediana 50 (loc) -> {}".format(np.random.normal(loc=50,size=(2,3))))

#y tambien un valor de scale el cual refiere a la desviacion estandar (el default es 1)
print("desviacion estandar 100 (scale), mediana 50 (loc) -> {}".format(np.random.normal(loc=50,scale=100,size=10)))

datox,datoy = np.random.randn(2, 100)
fig, ax = plt.subplots(1, 1)
graficar(ax, datox, datoy, {'marker': '*'})
```

```python
#trato de graficar el problema 01 de TSP_completo.txt:
'''
el archivo de entrada contiene:
5;
5;15;17;7;6;19;20;7;21;5;

Lo que significa que los costos entre ciudades (matriz simétrica):
        Ciudad1 Ciudad2 Ciudad3 Ciudad4 Ciudad5
Ciudad1     -      5      15      17       7
Ciudad2     -      -       6      19      20
Ciudad3     -      -       -       7      21
Ciudad4     -      -       -       -       5
Ciudad5     -      -       -       -       -

El procedimiento que voy a realizar para graficar los puntos (los cuales desconozco) son:
1) tomar un punto aleatorio de la ciudad1 -> por ejemplo el centro de un gráfico cuadrado de lado costo máximo*3
2) dibujar círculos con radio costos desde ciudad1 hacia las demas (primer fila de la matriz)
3) tomar un punto aleatorio del circulo correspondiente a ciudad2
4) calcular los dos puntos correspondientes a la distancia desde ciudad2 hacia ciudad3 (igual al costo hacia dicha ciudad).
5) tomar desde el punto encontrado (desde ciudad2 hasta ciudad3) y medir una distancia hacia ciudad4 igual al costo desde ciudad3
6) tomar desde el punto encontrado (desde ciudad3 hasta ciudad4) y medir una distancia hacia ciudad5 igual al costo desde ciudad4
'''

%matplotlib inline

import matplotlib.pyplot as plt
import random as rnd
from math import pi, sin, cos
from intersect import intersection

limiteX = 800/10
limiteY = 600/10
semilla = 72

alfa = np.linspace(0, 2*pi, 100)
iRand = rnd.Random(semilla)
fig, ax = plt.subplots(1, 1)

ax.set_xlim(0,limiteX)
ax.set_ylim(0,limiteY)
ax.grid(True)
ax.set_title('grafico introductorio')

# 1) y 2)
## valores de x e y de la ciudad 1
c1_x = limiteX/2
c1_y = limiteY/2

### una opcion es esto: #ciudad = plt.Circle((limiteX/2,limiteY/2), 5, color='r', fill=False) y luego graficar: ax.add_artist(ciudad3) y otra opcion es:
## funcion circulo de la ciudad 2
c1_c2_circ_x = limiteX/2+5*np.cos(alfa)
c1_c2_circ_y = limiteY/2+5*np.sin(alfa)
## funcion circulo de la ciudad 3
c1_c3_circ_x = limiteX/2+15*np.cos(alfa)
c1_c3_circ_y = limiteY/2+15*np.sin(alfa)
## funcion circulo de la ciudad 4
c1_c4_circ_x = limiteX/2+17*np.cos(alfa)
c1_c4_circ_y = limiteY/2+17*np.sin(alfa)
## funcion circulo de la ciudad 5
c1_c5_circ_x = limiteX/2+7*np.cos(alfa)
c1_c5_circ_y = limiteY/2+7*np.sin(alfa)

# 3) para tomar un punto aleatorio del circulo voy a calcular (al azar) un angulo entre 0 a 360 y situar el punto en dicho angulo del circulo
alfa2 = alfa[iRand.randint(0,len(alfa)-1)]
c2_x = limiteX/2+5*np.cos(alfa2)
c2_y = limiteY/2+5*np.sin(alfa2)

# 4) se calculan dos posibles puntos de la ciudad3 (desde la ciudad2) cumpliendo con el costo desde la ciudad1
# para ello se dibuja el circulo desde ciudad2 a ciudad3
c2_c3_circ_x = c2_x+6*np.cos(alfa)
c2_c3_circ_y = c2_y+6*np.sin(alfa)

c3_xs, c3_ys = intersection(c1_c2_circ_x, c1_c2_circ_y, c2_c3_circ_x, c2_c3_circ_y)

print("interseccion ciudad2 a ciudad3 -> {},{}".format(c3_xs,c3_ys))
# se toma uno de los dos puntos (al azar)
i_azar = iRand.randint(0,len(c3_xs)-1)
c3_x = c3_xs[i_azar]
c3_y = c3_ys[i_azar]

#grafico de los circulos de las ciudades
ax.plot(c1_c2_circ_x, c1_c2_circ_y, c='r', label='c1 a c2')
ax.plot(c1_c3_circ_x, c1_c3_circ_y, c='g', label='c1 a c3')
ax.plot(c1_c4_circ_x, c1_c4_circ_y, c='b', label='c1 a c4')
ax.plot(c1_c5_circ_x, c1_c5_circ_y, c='y', label='c1 a c5')
ax.plot(c2_c3_circ_x, c2_c3_circ_y, c='k', label='c2 a c3')

#ubicacion de las ciudades
print("ciudad1 -> x : {}, y: {}".format(c1_x,c1_y))
print("ciudad2 -> x : {}, y: {}".format(c2_x,c2_y))

#grafico de las ciudades
ax.plot(c1_x, c1_y, '*k', label='ciudad1')
ax.plot(c2_x, c2_y, '*r', label='ciudad2')
ax.plot(c3_x, c3_y, '*g', label='ciudad3')
#ax.plot(c4_x, c4_y, '*b', label='ciudad4')
#ax.plot(c5_x, c5_y, '*y', label='ciudad5')
ax.legend(loc=2)
```

```python
'''
PRUEBA 2

Algoritmo A*
1) Cree un grafo de búsqueda G y una lista ABIERTA y ponga a s en cada uno de ellos.
2) Cree una lista vacía CERRADA.
3) Si ABIERTA está vacía retorne con Falla.
4) Remueva el 1er nodo = n de ABIERTA y colóquela en CERRADA.
5) Si n = Meta retorne con Éxito.
6) Genere los M sucesores de n que no sean sus ancestros y póngalos en G.
7) Establezca punteros a n desde sus descendientes que no estén ni en ABIERTA ni en CERRADA, agregue los M sucesores a la lista ABIERTA y en caso que ya estuvieran redireccione los punteros si es un camino más corto.
8) Reordene ABIERTA.
9) Vaya al punto 3

'''
def debuggear(listAbierta,listCerrada):
    print("listAbierta: {}, listCerrada: {}".format(listAbierta,listCerrada))
    return

# esta funcion devuelve la ciudad de menor costo de transporte desde la ciudad variable <inicio>
def calcMenCost(cantCiu,mat_costCiu,inicio,listExceptuadas):
    listExceptuadas.append(inicio)
    vec_costCiu_aux = mat_costCiu[inicio-1]
    #print("vec cost Ciu: {}".format(vec_costCiu_aux))
    menor = -1
    dest = -1
    a = 0
    while (menor == -1 and dest == -1):
        if (a+1) not in listExceptuadas:
            menor = vec_costCiu_aux[a]
            dest = (a+1)
        else:
            a += 1
    for a in range(cantCiu-1):
        if (a+1) in listExceptuadas:
            continue
        aux = vec_costCiu_aux[a]
        if menor > aux and aux != 0:
            menor = aux
            dest = (a+1)
    return dest,listExceptuadas

import networkx as nx
import matplotlib as plt
import numpy as np

cantCiu = 5
vec_costCiu = [5,15,17,7,6,19,20,7,21,5]
mat_costCiu = [
    [ 0, 5,15,17, 7],
    [ 5, 0, 6,19,20],
    [15, 6, 0, 7,21],
    [17,19, 7, 0, 5],
    [ 7,20,21, 5, 0]
]

## Modelo mi algoritmo A* de forma tal que cada nodo del grafo es un recorrido (solución) de todas las ciudades
## en primera instancia puedo sacar un primer recorrido
## teniendo en cuenta (desde la ciudad que se tome como inicio) cual es la ciudad a la que el costo de moverse es menor
## si se dieran que dos ciudades destino tuviesen el mismo costo entonces se tomará alguna
## con esto puedo formular una primer solución
## ahora bien, para formular modificaciones al recorrido yo propongo -> conmutar de a pares solo 1 a la vez y calcular una predicción del costo total
## 

#1) Se crea un grafo de busqueda G y una lista ABIERTA

# creacion del grafo de busqueda G
grafoG = nx.Graph()

# generacion de la primer solucion base (se toma el menor de los costos partiendo de cualquier ciudad, en este caso de la ciudad 1)
inicio = 1
primRecorrido = [inicio]
listExceptuadas = []
for a in range(cantCiu-1):
    dest,listExceptuadas = calcMenCost(cantCiu,mat_costCiu,inicio,listExceptuadas) ##TODO: que pasa si hay mas de un destido de menor costo? ¿cual tomar?
    primRecorrido.append(dest)
    inicio = dest
    #print("iteracion a:{}".format(a))
    #print("dest: {}".format(dest))
    #print("primRecorrido:{}".format(primRecorrido))
    #print("listExceptuadas:{}".format(listExceptuadas))    
#print("primRecorrido: {}".format(primRecorrido))

# creacion lista ABIERTA
listAbierta = []

#los recorridos deben ser tuplas, no serán modificados a menos que se pase por un proceso de generacion de recorridos
listAbierta.append(tuple(primRecorrido))

print("listAbierta: {}".format(listAbierta))

# gerenacion de variaciones del primer recorrido con un metodo de:
# a) conmutación de pares de ciudades (como lista circular)
# b) simil cruza algoritmos geneticos -> ordered crossover
# c) tomar una ciudad y moverla 2 posiciones hacia adelante (a la izquierda) o hacia atras (a la derecha)
# d) idem anterior 3 posiciones
# e) doble conmutación a) en el mismo paso
#    me parece muy caótico pero podría destrabar alguna situación. Se puede dar el caso que haga lo mismo que en c) (si se conmutan dos pares que comparten 1 ciudad)
# f) 

# Segun a) ==================
#

#
# Fin a) ====================

# Segun b) ==================
# 

#
# Fin b) ====================

# Segun c) ==================
# 

#
# Fin c) ====================

# se agrega al grafo toda la listAbierta
#grafoG.add_node(listAbierta)
grafoG.add_nodes_from(listAbierta)

# 2) Se crea una lista vacía CERRADA
listCerrada = []

# 3) si ABIERTA está vacío -> ERROR
if len(listAbierta) == 0:
    print("ERROR")
    #exit(0)

# 4) Se remueve el primer nodo de ABIERTA y se coloca en CERRADA
listCerrada.append(listAbierta[0])
listAbierta = listAbierta[1:]

# 5) Si n = meta se retorna con exito
debuggear(listAbierta,listCerrada)

# 6) Se generan los M sucesores de n que no son sus ancestros y se ponen en G
# generar variaciones en la solucion n

# comprobar que esas variaciones no existen aguas arriba en el grafo, si existen descartarlas
# poner en G los nuevos sucesores
```
